```python
from os import walk

filenames = next(walk("/content/drive/MyDrive/Colab stylegan3"), (None, None, []))[2]  # [] if no file
filenames
```

```python
import cv2
from google.colab.patches import cv2_imshow

size = (300,300)

def resize_all(down_size, path):
  for f in filenames:
    dir = "{}/{}".format(path, f)
    image = cv2.imread(dir)
    if image is not None:
      image = cv2.resize(image, down_size)
      cv2.imwrite(dir, image)

resize_all(size, "/content/drive/MyDrive/Colab stylegan3")
print("done")
```
